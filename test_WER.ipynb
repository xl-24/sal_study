```python
!pip install -q jiwer
!pip install -q evaluate


from jiwer import wer
import evaluate

metric = evaluate.load("wer")
references  = ["praha radar oscar kilo k uniform november direct to inbound to"]
predictions = ["radar oscar kilo kilo uniform november dobry den direct to inbound to vladivostok"]


wer_score = metric.compute(predictions=predictions, references=references)
print("WER1:", wer_score)
sample_wer = 100 * wer(references, predictions)
print("WER2:", sample_wer)
```
